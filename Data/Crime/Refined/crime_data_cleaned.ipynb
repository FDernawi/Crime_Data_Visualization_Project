```python
import pandas as pd
import numpy as nm
```

```python
# Creating a file path to conduct manipulation
file_path = '/Users/faisaldernawi/Desktop/Training_And_Development/GW_Bootcamp/Homework/Project_3/Data/Crime/Unrefined/Crime_Incidents_in_2020.csv'
```

```python
# Loading the CSV file into a Pandas Dataframe
crime_data = pd.read_csv(file_path)
```

```python
# Displaying the Dataframe
crime_data.head()
```

```python
# Creating group of columns to keep
columns_to_keep = ['OFFENSE', 'LATITUDE', 'LONGITUDE', 'CENSUS_TRACT']
```

```python
# Dropping the rest of the columns that are redundant or not needed
crime_data_cleaned = crime_data[columns_to_keep]
```

```python
# Displaying the dataframe 
crime_data_cleaned
```

```python
# locking and reverting numeric values 
crime_data_cleaned.loc[:, 'CENSUS_TRACT'] = pd.to_numeric(crime_data_cleaned['CENSUS_TRACT'], errors='coerce')
```

```python
# Creating a copy of dataframe to modify
crime_data_cleaned = crime_data_cleaned.copy()
crime_data_cleaned['CENSUS_TRACT'] = pd.to_numeric(crime_data_cleaned['CENSUS_TRACT'], errors='coerce')
```

```python
# Dropping values to ensure alignment
crime_data_cleaned = crime_data_cleaned.dropna(subset=['CENSUS_TRACT']).copy()
crime_data_cleaned['CENSUS_TRACT'] = pd.to_numeric(crime_data_cleaned['CENSUS_TRACT'], errors='coerce')
```

```python
# Changing decimal place
crime_data_cleaned['TRACT_DIVIDED'] = crime_data_cleaned['CENSUS_TRACT'] / 10

# Display the first few rows to verify
print(crime_data_cleaned[['CENSUS_TRACT', 'TRACT_DIVIDED']].head())
```

```python
crime_data_cleaned
```

```python
# Dropping column and renaming other column
# Drop the original 'CENSUS_TRACT' column
crime_data_cleaned.drop(columns=['CENSUS_TRACT'], inplace=True)

# Rename the 'TRACT_DIVIDED' column to 'CENSUS_TRACT'
crime_data_cleaned.rename(columns={'TRACT_DIVIDED': 'CENSUS_TRACT'}, inplace=True)

crime_data_cleaned
```

```python
# Reordering the columns
crime_data_cleaned = crime_data_cleaned[['CENSUS_TRACT', 'OFFENSE', 'LATITUDE', 'LONGITUDE']]

# Display the first few rows to verify
crime_data_cleaned
```

```python
# Changing column name so that merging is easier
crime_data_cleaned.rename(columns={'CENSUS_TRACT': 'tract','OFFENSE': 'offense', 'LATITUDE': 'lat', 'LONGITUDE': 'lon'}, inplace=True)
crime_data_cleaned
```

```python
# Removal of decimals to ensure join
crime_data_cleaned['tract_no_decimal'] = crime_data_cleaned['tract'].apply(lambda x: int(float(x)))
crime_data_cleaned
```

```python
# Bins creation
crime_data_cleaned['BIN'] = crime_data_cleaned['tract_no_decimal'].apply(lambda x: min(int(float(x)) // 100, 18) + 1)
crime_data_cleaned
```

```python
# Reordering bins
crime_data_sorted = crime_data_cleaned.sort_values(by='BIN')
crime_data_sorted
```
