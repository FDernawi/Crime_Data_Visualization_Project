```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/Users/faisaldernawi/Desktop/Training_And_Development/GW_Bootcamp/Homework/Project_3/Data/Grouped Data/grouped_data_output.csv')

# Plot scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(data['total_pop'], data['crime_count'], 
                      c=data['crime_count'], cmap='coolwarm', s=80, alpha=0.7, edgecolors='k')

# Add color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Crime Count')

# Title and labels
plt.title('Crime Count vs. Total Population by Census Tract')
plt.xlabel('Total Population')
plt.ylabel('Crime Count')
plt.grid(True, linestyle='--', alpha=0.6)

# Annotate Top 5 Tracts with Highest Crime
top_tracts = data.nlargest(5, 'crime_count')
for i, row in top_tracts.iterrows():
    plt.annotate(f"Tract {int(row['tract'])}", 
                 (row['total_pop'], row['crime_count']),
                 textcoords="offset points",
                 xytext=(5, 5), ha='center', fontsize=10, color='darkred')

plt.show()
```
